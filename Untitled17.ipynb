```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

spark = SparkSession.builder \
    .appName("BigDataProject") \
    .getOrCreate()

data = [
    (1, "Laptop", "Electronics", 50000, 1),
    (2, "Mouse", "Electronics", 1000, 2),
    (3, "Table", "Furniture", 5000, 1),
    (4, "Phone", "Electronics", 20000, 1),
    (5, "Chair", "Furniture", 3000, 4)
]

schema = StructType([
    StructField("ID", IntegerType(), True),
    StructField("Product", StringType(), True),
    StructField("Category", StringType(), True),
    StructField("Price", IntegerType(), True),
    StructField("Quantity", IntegerType(), True)
])

df = spark.createDataFrame(data, schema)

df.show()

from pyspark.sql.functions import col
df_revenue = df.withColumn("Revenue", col("Price") * col("Quantity"))

result = df_revenue.groupBy("Category").sum("Revenue")

result.show()

spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
from pyspark.sql.functions import col

# Initialize the session
spark = SparkSession.builder \
    .appName("BigDataProject") \
    .getOrCreate()
```

```python
# Create the raw data
data = [
    (1, "Laptop", "Electronics", 50000, 1),
    (2, "Mouse", "Electronics", 1000, 2),
    (3, "Table", "Furniture", 5000, 1),
    (4, "Phone", "Electronics", 20000, 1),
    (5, "Chair", "Furniture", 3000, 4)
]

# Define the structure (Schema)
schema = StructType([
    StructField("ID", IntegerType(), True),
    StructField("Product", StringType(), True),
    StructField("Category", StringType(), True),
    StructField("Price", IntegerType(), True),
    StructField("Quantity", IntegerType(), True)
])

# Create the DataFrame
df = spark.createDataFrame(data, schema)
```

```python
# Step A: Show the raw data
print("Raw Sales Data:")
df.show()

# Step B: Calculate Revenue for each item
df_revenue = df.withColumn("Revenue", col("Price") * col("Quantity"))

# Step C: Group by Category to see total revenue
result = df_revenue.groupBy("Category").sum("Revenue")

# Step D: Final Output
print("Final Category-wise Revenue Analysis:")
result.show()

# Step E: Clean up
spark.stop()
```
